```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define CSV URL
csv_url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download'
# Read the csv file from the URL
df = pd.read_csv(csv_url)
```

```python
# Extract the domain from the 'email' column
df['domain'] = df['Email'].str.split('@').str[1]
# Count occurances of each unique domain value in 'domain' column
domain_counts = df['domain'].value_counts()
# Extract counts for 'domain_counts' to use as 'y' for pie chart
y = domain_counts.values  
# Extract domain names from 'domain_counts' to use as labels for the pie chart 
label = domain_counts.index  
```

```python
# Define plot size
plt.figure(figsize=(12, 8))
# Set custom colours
colors = ['#FFB6C1', '#87CEFA', '#90EE90']
# Create a pie chart using counts, displaying each slices percentage 
# With custom text and wedge style.
plt.pie(
    y,
    labels=label,
    autopct='%1.1f%%',
    colors=colors,
    textprops={'fontsize': 14, 'color': 'black'},  # Slightly smaller font for better clarity
    wedgeprops={'linewidth': 1.5, 'edgecolor': 'gray'}
)
# Add and set size of chart title 
plt.title('Email Domain Distribution', fontsize=24, fontweight='bold')
# Equal aspect ratio to ensure the pie chart is a circle
plt.axis('equal') 
# Show pie chart
plt.show()
```
