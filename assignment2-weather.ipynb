```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
# Define CSV URL
csv_url = 'https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv'
# Read the csv file from the URL
df = pd.read_csv(csv_url)
```

[Reference: How to read a CSV from a URL using Pandas?](https://www.askpython.com/python-modules/pandas/read-csv-from-url-pandas)

```python
# Define the relevant columns for the graph
# Temprature
temperature = df['dryBulbTemperature_Celsius']
# Time
# Convert 'reportStartDateTime' to datetime
time = pd.to_datetime(df['reportStartDateTime']) 
```

```python
# Define plot size
plt.figure(figsize=(12, 6))
# Create a line plot using time and temperature.
# Define color ,line width ,label and add a marker.
plt.plot(time, temperature, color='lightblue', linewidth=2, label='Temperature',marker='o', markerfacecolor='darkblue',markersize=5)
# Add titles ,labels ,grid and legend 
plt.title('Temperature Over Time', fontsize=16, loc='center')
plt.xlabel('Date and Time', fontsize=14)
plt.ylabel('Temperature (C)', fontsize=14)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Set date formatting on x-axis
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d %H:%M'))
plt.xticks(rotation=45,ha='right')

plt.tight_layout()

# Show the plot
plt.show()
```
