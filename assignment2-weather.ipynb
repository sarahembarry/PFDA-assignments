```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define CSV URL
csv_url = 'https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv'
# Read the csv file from the URL
df = pd.read_csv(csv_url)
```

```python
# Define the relevant columns for the graph
# Temprature
temperature = df['dryBulbTemperature_Celsius']
# Time
# Convert 'reportStartDateTime' to datetime
time = pd.to_datetime(df['reportStartDateTime']) 
```

```python
# Define plot size
plt.figure(figsize=(10, 6))
# Create a line plot using time and temperature.
# Define color ,line width ,label and add a marker.
plt.plot(time, temperature, color='green', linewidth=2, label='Temperature',marker='o', markerfacecolor='blue')
# Add titles ,labels ,grid and legend 
plt.title('Temprature over time' ,fontsize=20)
plt.xlabel('Time', fontsize=13)
plt.ylabel('Temperature (°C)', fontsize=13)
plt.grid(color = 'blue', linestyle = '--', linewidth = 0.5)
plt.legend()
# Show the plot
plt.show()
```
